# Polygonal Approximation of Objects' Contours

This notebook compares and visualizes three classic polygonal simplification algorithms applied to noisy contour data:
- **Douglas-Peucker (RDP)**
- **Visvalingam-Whyatt (VW)**
- **Minimum Perimeter Polygon (MPP)**

Each algorithm attempts to reduce the number of vertices in a contour while preserving its overall shape within a given error threshold or structural constraint.


```python
%matplotlib inline

from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np
from typing import List, Tuple, Optional, Union
import os
```

### Preprocessing Functions
- `border_trace_image`: A classic 8-connected Moore-neighbor following algorithm is used to trace the first
black (foreground) contour in a binary image.
- `process_folder`: Iteratively applies the previous function on each image in the input folder.

```python
def border_trace_image(input_path: str, output_path: str) -> None:
    """
    Trace the first black (0) contour in a grayscale image using 8-connected Moore
    neighbor following. The traced border is drawn in white (255) on a black background.
    """
    # 8-connected neighbor offsets (row, col)
    di = [ 0, -1, -1, -1,  0,  1,  1,  1 ]
    dj = [ 1,  1,  0, -1, -1, -1,  0,  1 ]

    # Load as grayscale
    src = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
    if src is None:
        print(f"Error: Could not read {input_path!r}")
        return

    rows, cols = src.shape
    dst = np.zeros_like(src, dtype=np.uint8)

    # Find the first black pixel
    p0: Optional[Tuple[int, int]] = None
    for i in range(rows):
        for j in range(cols):
            if src[i, j] == 0:
                p0 = (i, j)
                break
        if p0 is not None:
            break

    if p0 is None:
        print(f"Warning: No black pixels in {input_path!r}")
        return

    # Initialize
    p_curr = p0
    dir_ = 7           # starting search direction
    n = 0
    dst[p_curr] = 255  # mark the start
    first_pass = True
    p1: Optional[Tuple[int, int]] = None

    # Moore-neighbor border following
    while True:
        p_prev = p_curr
        n += 1

        # step one back from the arrival direction
        dir_ = (dir_ + 7) % 8 if dir_ % 2 == 0 else (dir_ + 6) % 8

        # find next black neighbor by rotating clockwise
        nr = p_prev[0] + di[dir_]
        nc = p_prev[1] + dj[dir_]
        while (
                rows > nr >= 0 != src[nr, nc] and
                0 <= nc < cols
        ):
            dir_ = (dir_ + 1) % 8
            nr = p_prev[0] + di[dir_]
            nc = p_prev[1] + dj[dir_]

        # if we’ve stepped off the image, bail out
        if not (0 <= nr < rows and 0 <= nc < cols):
            break

        p_curr = (nr, nc)
        dst[p_curr] = 255

        if first_pass:
            p1 = p_curr
            first_pass = False

        # stop once we’ve returned to the starting configuration
        if p1 is not None and p_curr == p1 and p_prev == p0 and n >= 2:
            break

    # Write result
    cv2.imwrite(output_path, dst)
```

```python
# Usage example
border_trace_image('files_border_tracing/horizontal_ellipse.bmp', 'output.bmp')
```

```python
def has_image_extension(filename: Union[str, Path]) -> bool:
    """
    Return True if filename ends with a supported image extension.
    Supports: .bmp, .png, .jpg, .jpeg (case-insensitive).
    """
    return str(filename).lower().endswith((
        ".bmp", ".png", ".jpg", ".jpeg"
    ))

def process_folder(input_folder: Union[str, Path],
                   output_folder: Union[str, Path]) -> None:
    """
    Walks through all files in `input_folder`, and for each supported image
    file, calls `border_trace_image(input_path, output_path)`, writing results
    to `output_folder` under the same filename.
    """
    input_path = Path(input_folder)
    output_path = Path(output_folder)

    if not input_path.is_dir():
        print(f"Error: Could not open input folder: {input_path}")
        return

    # Create output folder if it doesn't exist
    output_path.mkdir(parents=True, exist_ok=True)

    for entry in input_path.iterdir():
        if not entry.is_file():
            continue  # skip subdirectories, symlinks, etc.

        if has_image_extension(entry.name):
            in_file = entry
            out_file = output_path / entry.name
            print(f"Processing: {in_file}")
            border_trace_image(str(in_file), str(out_file))
```

### Helper functions
- `perpendicular_distance`: used by RDP and MPP algorithm to assess the error (the distance from the polyline to the actual contour)
- `triangle_area`: used by VW algorithm to rank points based on the area of the trinagle they describe

```python
def perpendicular_distance(
    pt: Tuple[int, int],
    line_start: Tuple[int, int],
    line_end: Tuple[int, int]
) -> float:
    x0, y0 = pt
    x1, y1 = line_start
    x2, y2 = line_end
    dx, dy = x2 - x1, y2 - y1
    if dx == 0 and dy == 0:
        return math.hypot(x0 - x1, y0 - y1)
    num = abs(dy * x0 - dx * y0 + x2 * y1 - y2 * x1)
    den = math.hypot(dx, dy)
    return num / den
```

```python
def triangle_area(p1, p2, p3):
    return abs((p1[0]*(p2[1]-p3[1]) + p2[0]*(p3[1]-p1[1]) + p3[0]*(p1[1]-p2[1])) / 2.0)
```

### 1. Douglas-Peucker (RDP)

This is a recursive algorithm that works by:
- Connecting the endpoints of a curve segment.
- Finding the point farthest from the segment.
- If this distance exceeds a threshold ε, the curve is split and processed recursively.

#### Key Properties:
- Preserves key geometric corners.
- Highly sensitive to ε: small ε retains more points; large ε leads to aggressive simplification.

```python
def douglas_peucker(pts: List[Tuple[int, int]], eps: float) -> List[Tuple[int, int]]:
    if len(pts) < 2:
        return pts[:]
    start, end = pts[0], pts[-1]
    dmax, index = 0.0, 0
    for i in range(1, len(pts) - 1):
        d = perpendicular_distance(pts[i], start, end)
        if d > dmax:
            dmax, index = d, i
    if dmax > eps:
        left = douglas_peucker(pts[:index+1], eps)
        right = douglas_peucker(pts[index:], eps)
        return left[:-1] + right
    else:
        return [start, end]
```

### 2. Visvalingam-Whyatt (VW)

VW ranks points based on the area of the triangle they form with their neighbors:
- Points with the smallest triangle area are the least important.
- Simplification proceeds by removing least-important points until a desired **ratio** of points is retained.

#### Key Properties:
- Smoother simplification than RDP.
- Excellent for generalizing organic shapes like coastlines or ellipses.
- Ratio-based control is more intuitive than area thresholds.

```python
import heapq
from typing import List, Tuple
import math

def visvalingam_whyatt(pts: List[Tuple[int, int]], ratio: float) -> List[Tuple[int, int]]:
    if len(pts) <= 2 or ratio >= 1.0:
        return pts[:]

    class PointData:
        def __init__(self, index, area):
            self.index = index
            self.area = area
        def __lt__(self, other):
            return self.area < other.area

    n = len(pts)
    target = max(2, int(n * ratio))  # ensure at least 2 points remain
    areas = [math.inf] * n
    heap = []

    for i in range(1, n - 1):
        areas[i] = triangle_area(pts[i - 1], pts[i], pts[i + 1])
        heapq.heappush(heap, PointData(i, areas[i]))

    removed = [False] * n
    remaining = n

    while remaining > target and heap:
        point = heapq.heappop(heap)
        i = point.index
        if removed[i]:
            continue
        removed[i] = True
        remaining -= 1
        for j in [i - 1, i + 1]:
            if 0 < j < n - 1 and not removed[j]:
                areas[j] = triangle_area(pts[j - 1], pts[j], pts[j + 1])
                heapq.heappush(heap, PointData(j, areas[j]))

    return [pt for i, pt in enumerate(pts) if not removed[i]]
```

### 3. Minimum Perimeter Polygon (MPP)

The MPP method iteratively removes points that:
- Reduce the total perimeter the most.
- Do not exceed a given **perpendicular distance error threshold** when approximated as a line segment between neighbors.

#### Key Properties:
- Geometrically motivated (perimeter efficiency).
- Balances shape retention with local smoothness.
- Naturally avoids over-simplifying curved regions.

```python
# Redefine MPP using perpendicular distance as error threshold
def minimum_perimeter_polygon(pts: List[Tuple[int, int]], eps: float) -> List[Tuple[int, int]]:
    def dist(a: Tuple[int, int], b: Tuple[int, int]) -> float:
        return math.hypot(a[0] - b[0], a[1] - b[1])

    pts = pts[:]
    while True:
        min_inc = float('inf')
        idx = -1
        for i in range(1, len(pts) - 1):
            p1, p2, p3 = pts[i - 1], pts[i], pts[i + 1]
            inc = dist(p1, p2) + dist(p2, p3) - dist(p1, p3)
            err = perpendicular_distance(p2, p1, p3)
            if inc < min_inc and err <= eps:
                min_inc = inc
                idx = i
        if idx == -1:
            break
        pts.pop(idx)
    return pts
```

```python
# Collect image files from directory
image_dir = "contour_imgs"
image_files = sorted([
    f for f in os.listdir(image_dir)
    if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))
])
if not image_files:
    raise FileNotFoundError("No image files found in 'contour_imgs/' directory.")

def interactive_display(image_index: int, epsilon: float, ratio: float):
    from IPython.display import clear_output
    import cv2
    import numpy as np
    import matplotlib.pyplot as plt
    import os

    clear_output(wait=True)

    image_path = os.path.join(image_dir, image_files[image_index])
    print(f"File: {image_files[image_index]} | Epsilon: {epsilon} | Ratio: {ratio}")

    src = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if src is None:
        print(f"Error: Cannot read image at {image_path}")
        return

    contours, _ = cv2.findContours(src, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    result = cv2.cvtColor(src, cv2.COLOR_GRAY2BGR)

    for cnt in contours:
        if len(cnt) < 3:
            continue
        pts = [tuple(pt[0]) for pt in cnt]

        dp_pts = douglas_peucker(pts, epsilon)
        mpp_pts = minimum_perimeter_polygon(pts, epsilon)
        vw_pts = visvalingam_whyatt(pts, ratio)

        dp_np = np.array(dp_pts, dtype=np.int32).reshape(-1, 1, 2)
        vw_np = np.array(vw_pts, dtype=np.int32).reshape(-1, 1, 2)
        mpp_np = np.array(mpp_pts, dtype=np.int32).reshape(-1, 1, 2)

        cv2.polylines(result, [dp_np], isClosed=False, color=(0, 0, 255), thickness=1)   # Red
        cv2.polylines(result, [vw_np], isClosed=False, color=(0, 255, 0), thickness=1)   # Green
        cv2.polylines(result, [mpp_np], isClosed=False, color=(255, 0, 0), thickness=1)  # Blue

    result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(6, 6))
    plt.imshow(result_rgb)
    plt.axis('off')
    plt.title(f"{image_files[image_index]}\nRed: DP (ε={epsilon}) | Green: VW (r={ratio}) | Blue: MPP (ε={epsilon})")
    plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display

image_slider = widgets.IntSlider(
    value=0,
    min=0,
    max=len(image_files) - 1,
    step=1,
    description='Image:'
)

epsilon_slider = widgets.FloatSlider(
    value=2.0, min=0.0, max=30.0, step=0.5, description='Epsilon:'
)

ratio_slider = widgets.FloatSlider(
    value=0.5, min=0.01, max=1.0, step=0.01, description='VW Ratio:'
)

interactive_widget = widgets.interactive_output(
    interactive_display,
    {
        'image_index': image_slider,
        'epsilon': epsilon_slider,
        'ratio': ratio_slider
    }
)

display(image_slider, epsilon_slider, ratio_slider, interactive_widget)
```

## Experimental Setup

- A synthetic noisy ellipse was generated using 100 sample points with added Gaussian noise.
- Each simplification algorithm was applied using common parameters:
  - **ε = 5.0** for RDP and MPP.
  - **Retention Ratio = 0.25** for VW.
- The results were visualized side by side to assess performance.

The plot shows how each method simplifies the original noisy contour.


```python
# Generate synthetic ellipse
theta = np.linspace(0, 2 * np.pi, 100)
a, b = 100, 50
x = a * np.cos(theta) + np.random.randn(100) * 2
y = b * np.sin(theta) + np.random.randn(100) * 2
ellipse_pts = list(zip(x.astype(int), y.astype(int)))

# Apply all three
epsilon = 5.0
ratio = 0.25
dp_pts = douglas_peucker(ellipse_pts, epsilon)
mpp_pts = minimum_perimeter_polygon(ellipse_pts, epsilon)
vw_pts = visvalingam_whyatt(ellipse_pts, ratio)

# Plot results
plt.figure(figsize=(8, 6))
plt.plot(*zip(*ellipse_pts), label="Original", color='gray', linewidth=1)
plt.plot(*zip(*dp_pts), label=f"Douglas-Peucker (ε={epsilon})", color='red', linewidth=1.5)
plt.plot(*zip(*vw_pts), label=f"Visvalingam-Whyatt (r={ratio})", color='green', linewidth=1.5)
plt.plot(*zip(*mpp_pts), label=f"MPP (ε={epsilon})", color='blue', linewidth=1.5)
plt.axis('equal')
plt.legend()
plt.title("Polygonal Approximation: DP, VW, MPP")
plt.grid(True)
plt.show()
```

```python
# Plot each simplified result in a separate subplot for clarity
fig, axs = plt.subplots(1, 4, figsize=(16, 4), constrained_layout=True)

# Original
axs[0].plot(*zip(*ellipse_pts), color='gray', linewidth=1)
axs[0].set_title(f"Original ({len(ellipse_pts)} pts)")
axs[0].axis('equal')
axs[0].axis('off')

# Douglas-Peucker
axs[1].plot(*zip(*dp_pts), color='red', linewidth=1.5)
axs[1].set_title(f"Douglas-Peucker (ε={epsilon})\n{len(dp_pts)} pts")
axs[1].axis('equal')
axs[1].axis('off')

# Visvalingam-Whyatt
axs[2].plot(*zip(*vw_pts), color='green', linewidth=1.5)
axs[2].set_title(f"VW (r={ratio})\n{len(vw_pts)} pts")
axs[2].axis('equal')
axs[2].axis('off')

# MPP
axs[3].plot(*zip(*mpp_pts), color='blue', linewidth=1.5)
axs[3].set_title(f"MPP (ε={epsilon})\n{len(mpp_pts)} pts")
axs[3].axis('equal')
axs[3].axis('off')

plt.suptitle("Polygonal Simplification Algorithms", fontsize=14)
plt.show()
```
